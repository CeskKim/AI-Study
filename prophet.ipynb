```python
# prophet quick start
import pandas as pd
from prophet import Prophet

# source data
df = pd.read_csv('https://raw.githubusercontent.com/facebook/prophet/main/examples/example_wp_log_peyton_manning.csv')
df.head()

# fit setting
m = Prophet()
m.fit(df)

future = m.make_future_dataframe(periods=365)
future.tail()
```

```python
# prophet model
from prophet import Prophet
bitcoin_stock = fdr.DataReader('BTC/KRW', '2022-12-01', '2023-02-14')
bitcoin_stock.tail()
bitcoin_stock['y'] = bitcoin_stock['Close']
bitcoin_stock['ds'] = bitcoin_stock.index

pr = Prophet()
pr.fit(bitcoin_stock)

# prophet prediction period
future = pr.make_future_dataframe(periods=60)
forecast = pr.predict(future)
forecast.tail()
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].iloc[-40:-20]
```
